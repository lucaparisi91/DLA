```python
import dla
import numpy as np
```

```python
shape=[100,100,100]
N=1000
rng=np.random.default_rng()

def randomConfiguration(N,shape):
    free=np.zeros( (N,len(shape) ))
    for d in range(len(shape)):
        free[:,d]=rng.integers(0,shape[d]-1,size=N)
    return free
free=randomConfiguration(N,shape)
cluster=randomConfiguration(100,shape)
sim=dla.dla(shape=shape,state=(free,cluster))
```

```python
%matplotlib notebook
dla.plot3d(sim.state)
```

```python
sim.advance( 1000)
```
