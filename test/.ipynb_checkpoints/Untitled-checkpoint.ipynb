```python
import numpy as np
from numpy.random import default_rng
import dla
import matplotlib.pylab as plt
```

```python
def plotCluster(x,y):

    fig = plt.figure()
    ax = fig.add_subplot(projection='3d')
    if x is not None:
        ax.scatter(x[:,0], x[:,1], x[:,2],color="blue" )
    ax.scatter(y[:,0], y[:,1], y[:,2],color="red" )
    return (fig)
```

```python
rng = default_rng(35)
L=10
x=rng.uniform(low=-L/2,high=L/2,size=(100,3))
y=rng.uniform(low=-L/2,high=L/2,size=(1,3))

alg=dla.dla( (x,y),sigma=0.02,radius=1)
```

```python
%matplotlib notebook
fig=plotCluster(x,y)
```

```python
alg.advance(10000)
```

```python
%matplotlib notebook
fig=plotCluster(alg.free,alg.cluster)
```
