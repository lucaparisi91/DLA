```python
import numpy as np
from numpy.random import default_rng
import dla
import matplotlib.pylab as plt
import tqdm
from importlib import reload
reload(dla)
```

```python
def plotCluster(x,y=None):
    fig = plt.figure()
    ax = fig.add_subplot(projection='3d')
    if x is not None:
        ax.scatter(x[:,0], x[:,1], x[:,2],color="green" )
    if y is not None:
        ax.scatter(y[:,0], y[:,1], y[:,2],color="red" )
    return (fig)
```

```python
rng = default_rng(39)
L=100
x=rng.uniform(low=-L/2,high=L/2,size=(1000,3))
y=rng.uniform(low=-L/2,high=L/2,size=(20,3))
y[:,2]=0
x[:,2]=0
alg=dla.dla( (x,y),sigma=[0.1,0.1,0],radius=1,lBox=[L,L,L])
```

```python
%matplotlib notebook
cluster=alg.cluster
free=alg.free
fig=plotCluster(free,cluster)
```

```python
for i in tqdm.tqdm(range(10)):
    alg.advance(1000)
```
